```python
import pandas as pd
import plotly.io as pio
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('metrics.csv')
```

```python
df
```

```python
fig = make_subplots(specs=[[{"secondary_x": True}]])
fig.add_trace(
    go.scatter.Line(x=df["step"], y=df["episode_reward"], name="xaxis data"),
    secondary_x=False,
)
fig.add_trace(
    go.scatter.Line(x=df["episode"], name="xaxis2 data"),
    secondary_x=True,
)

fig.update_layout(
    title_text="The Episode Reward over Time(training step)"
)

fig.show()
```

```python
fig = px.line(df, x="step", y="episode_reward", title="The Episode Reward over Time (in training step of model)")
fig.show()
```

```python
fig = px.line(df, x="step", y="train_loss", title="The Episode Reward over Time")
fig.show()
```
